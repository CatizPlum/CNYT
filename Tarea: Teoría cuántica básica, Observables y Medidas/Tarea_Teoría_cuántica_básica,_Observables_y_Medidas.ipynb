# Simule el primer sistema cuántico descrito en la sección 4.1.



```python
import numpy as np

class QuantumSystem:

    def __init__(self, ket):
        """
        ket: vector de estado (lista o array de números complejos)
        """
        self.ket = np.array(ket, dtype=complex)
        self.normalize()

    def normalize(self):
        """Normaliza el vector de estado"""
        norm = np.linalg.norm(self.ket)
        if norm == 0:
            raise ValueError("El vector no puede ser cero")
        self.ket = self.ket / norm

    def probability_position(self, i):
        """
        Calcula la probabilidad de encontrar la partícula en la posición i
        """
        if i < 0 or i >= len(self.ket):
            raise IndexError("Posición fuera del rango")

        prob = np.abs(self.ket[i])**2
        return prob

    def transition_probability(self, other_ket):
        """
        Calcula la probabilidad de transición a otro estado
        """
        other_ket = np.array(other_ket, dtype=complex)

        # Normalizar el otro estado
        other_ket = other_ket / np.linalg.norm(other_ket)

        # Producto interno (bra-ket)
        amplitude = np.vdot(other_ket, self.ket)

        # Probabilidad
        return np.abs(amplitude)**2
```

## EJEMPLO DE PROBABILIDAD EN UNA POSICIÓN EN PARTICULAR
### Estado inicial

```python
ket = [1/np.sqrt(2), 1j/np.sqrt(2)]

qs = QuantumSystem(ket)

# Probabilidad de posición
print("Probabilidad en posición 0:", qs.probability_position(0))
print("Probabilidad en posición 1:", qs.probability_position(1))
```

## EJEMPLO DE PASAR DE UN VECTOR A OTRO

```python
ket1 = [1/np.sqrt(2), 1j/np.sqrt(2)]
ket2 = [1, 0]

qs = QuantumSystem(ket1)

prob = qs.transition_probability(ket2)

print("Probabilidad de transición:", prob)
```

## 1. Amplitud de transición. El sistema puede recibir dos vectores y calcular la probabilidad de transitar de el uno al otro después de hacer la observación.

```python
import numpy as np

def normalize(ket):
    ket = np.array(ket, dtype=complex)
    norm = np.linalg.norm(ket)
    if norm == 0:
        raise ValueError("El vector no puede ser cero")
    return ket / norm


def transition_amplitude(ket1, ket2):
    """
    Calcula la amplitud de transición <ket2 | ket1>
    """
    ket1 = normalize(ket1)
    ket2 = normalize(ket2)

    return np.vdot(ket2, ket1)


def transition_probability(ket1, ket2):
    """
    Calcula la probabilidad de transición entre dos estados
    """
    amplitude = transition_amplitude(ket1, ket2)
    return np.abs(amplitude)**2

# EJEMPLO
ket1 = [1/np.sqrt(2), 1j/np.sqrt(2)]
ket2 = [1, 0]

amp = transition_amplitude(ket1, ket2)
prob = transition_probability(ket1, ket2)

print("Amplitud de transición:", amp)
print("Probabilidad de transición:", prob)
```

## 2. Ahora con una matriz que describa un observable y un vector ket, el sistema revisa que la matriz sea hermitiana, y si lo es, calcula la media y la varianza del observable en el estado dado.

```python
import numpy as np

def normalize(ket):
    ket = np.array(ket, dtype=complex)
    norm = np.linalg.norm(ket)
    if norm == 0:
        raise ValueError("El vector no puede ser cero")
    return ket / norm


def is_hermitian(A):
    A = np.array(A, dtype=complex)
    return np.allclose(A, A.conj().T)


def check_dimensions(A, ket):
    if A.shape[0] != A.shape[1]:
        raise ValueError("La matriz debe ser cuadrada")
    if A.shape[1] != len(ket):
        raise ValueError("Dimensiones incompatibles entre A y el ket")


def expected_value(A, ket):
    ket = normalize(ket)
    return np.vdot(ket, A @ ket)


def variance(A, ket):
    ket = normalize(ket)
    mu = expected_value(A, ket)
    I = np.eye(A.shape[0])

    delta = A - mu * I

    return np.real(np.vdot(ket, delta @ delta @ ket))


def observable_analysis(A, ket):
    A = np.array(A, dtype=complex)
    ket = np.array(ket, dtype=complex)

    check_dimensions(A, ket)

    if not is_hermitian(A):
        raise ValueError("La matriz NO es hermitiana")

    mu = expected_value(A, ket)
    var = variance(A, ket)

    return mu, var
```

## EJEMPLO

```python
# Observable (hermitiano)
A = np.array([[2, 1-1j],
              [1+1j, 3]])

# Estado
ket = [1/np.sqrt(2), 1/np.sqrt(2)]

mu, var = observable_analysis(A, ket)

print("¿Es hermitiana?:", is_hermitian(A))
print("Valor esperado:", mu)
print("Varianza:", var)
```

## 3. El sistema calcula los valores propios del observable y la probabilidad de que el sistema transite a alguno de los vectores propios después de la observación.

```python
import numpy as np

def normalize(ket):
    ket = np.array(ket, dtype=complex)
    norm = np.linalg.norm(ket)
    if norm == 0:
        raise ValueError("El vector no puede ser cero")
    return ket / norm


def is_hermitian(A):
    return np.allclose(A, A.conj().T)


def check_dimensions(A, ket):
    if A.shape[0] != A.shape[1]:
        raise ValueError("La matriz debe ser cuadrada")
    if A.shape[1] != len(ket):
        raise ValueError("Dimensiones incompatibles")

def eigen_analysis(A, ket):
    A = np.array(A, dtype=complex)
    ket = normalize(ket)

    check_dimensions(A, ket)

    if not is_hermitian(A):
        raise ValueError("La matriz NO es hermitiana")

    # Eigenvalores y eigenvectores
    eigenvalues, eigenvectors = np.linalg.eig(A)

    results = []

    for i in range(len(eigenvalues)):
        v = eigenvectors[:, i]
        v = normalize(v)

        prob = np.abs(np.vdot(v, ket))**2

        results.append({
            "eigenvalue": eigenvalues[i],
            "eigenvector": v,
            "probability": prob
        })

    return results
```

## EJEMPLO

```python
A = np.array([[2, 1-1j],
              [1+1j, 3]])

ket = [1/np.sqrt(2), 1/np.sqrt(2)]

results = eigen_analysis(A, ket)

for r in results:
    print("Valor propio:", r["eigenvalue"])
    print("Vector propio:", r["eigenvector"])
    print("Probabilidad:", r["probability"])
    print("-" * 40)
```

## 4. Se considera la dinámica del sistema. Ahora con una serie de matrices Un el sistema calcula el estado final a partir de un estado inicial.

```python
import numpy as np

def normalize(ket):
    ket = np.array(ket, dtype=complex)
    norm = np.linalg.norm(ket)
    if norm == 0:
        raise ValueError("El vector no puede ser cero")
    return ket / norm


def is_unitary(U):
    U = np.array(U, dtype=complex)
    I = np.eye(U.shape[0])
    return np.allclose(U.conj().T @ U, I)


def check_unitary_list(operators):
    for i, U in enumerate(operators):
        if not is_unitary(U):
            raise ValueError(f"La matriz U{i} NO es unitaria")


def check_dimensions(U, ket):
    if U.shape[1] != len(ket):
        raise ValueError("Dimensiones incompatibles")


def evolve_system(initial_ket, operators):
    state = normalize(initial_ket)

    check_unitary_list(operators)

    for U in operators:
        U = np.array(U, dtype=complex)
        check_dimensions(U, state)
        state = U @ state

    return state
```

## EJEMPLO

```python
# Estado inicial
ket = [1, 0]

# Matriz unitaria
U1 = np.array([[0, 1],
               [1, 0]])

U2 = np.array([[1, 0],
               [0, -1]])

final_state = evolve_system(ket, [U1, U2])

print("Estado final:", final_state)
```

# 4.5.2


En mecánica cuántica, el estado de una partícula con spin puede representarse como un vector en un espacio de dimensión 2:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$

donde $\alpha, \beta \in \mathbb{C}$ y $|\alpha|^2 + |\beta|^2 = 1$.


### Sistema de dos partículas

Cuando se consideran dos partículas, el espacio de estados total se construye mediante el **producto tensorial** de los espacios individuales:

$$
|\psi\rangle = c_{00}|00\rangle + c_{01}|01\rangle + c_{10}|10\rangle + c_{11}|11\rangle
$$

Este sistema vive en un espacio de dimensión:

$$
2^2 = 4
$$


### Generalización a $n$ partículas

Para un sistema de $n$ partículas con spin, el estado general se expresa como:

$$
|\psi\rangle = \sum_{i=0}^{2^n - 1} c_i |i\rangle
$$

donde cada $|i\rangle$ representa una configuración binaria de los spins.

El espacio de estados tiene dimensión:

$$
2^n
$$


### Implicación fundamental

El crecimiento exponencial del espacio de estados implica que:

- Un sistema con pocas partículas ya tiene una enorme cantidad de información
- Esto es la base de los **registros cuánticos (qubits)**
- Es lo que permite la ventaja potencial de la computación cuántica


### Interpretación física

Cada coeficiente $c_i$ representa la amplitud de probabilidad de encontrar el sistema en una configuración específica.

Sin embargo, el sistema no está en un solo estado, sino en una **superposición** de todos ellos, lo que implica que la información está distribuida globalmente en el sistema.

## Exercise 4.5.3 — Separabilidad y entrelazamiento

Considérese el estado:

$$
|\phi\rangle = |x0\rangle \otimes |y1\rangle + |x1\rangle \otimes |y1\rangle
$$


### Reescritura del estado

Factorizando términos comunes:

$$
|\phi\rangle = (|x0\rangle + |x1\rangle) \otimes |y1\rangle
$$


### Definición de separabilidad

Un estado es **separable** si puede escribirse como:

$$
|\psi\rangle = |\psi_1\rangle \otimes |\psi_2\rangle
$$

En este caso, claramente:

$$
|\psi_1\rangle = |x0\rangle + |x1\rangle
$$

$$
|\psi_2\rangle = |y1\rangle
$$


### Conclusión

El estado es **separable**, ya que puede expresarse como un producto tensorial de estados individuales.


### Interpretación física

Esto implica que:

- No existe entrelazamiento entre las partículas
- El estado de una partícula no depende del estado de la otra
- Las mediciones en una partícula no afectan instantáneamente a la otra


### Comparación con estados entrelazados

Un estado entrelazado típico sería:

$$
|\psi\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)
$$

Este estado **no puede factorizarse**, lo que implica:

- Correlaciones cuánticas fuertes
- Dependencia entre partículas
- Base de fenómenos como teleportación cuántica


### Conclusión general

La separabilidad distingue sistemas clásicos de sistemas cuánticos profundamente correlacionados. Mientras los estados separables pueden entenderse como combinaciones independientes, los estados entrelazados representan una propiedad intrínsecamente cuántica sin análogo clásico.
